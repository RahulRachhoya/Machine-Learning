Cross Product of vectors

```python
a = np.array([2,3])
b = np.array([4,4])

a_cross_b = np.cross(a,b)
```

```python
print(a_cross_b)
# To Calculate the output 
# a*b =  | i j k |  = i(3.0-0.4)-j(2.0-0.4)+k(2.4+3.4) = i(0-0) + j(0-0) + k(8-12) = {(0)i-(0)j+(-4)k}
#        | 2 3 0 |
#        | 4 4 0 |
# Note where . = represent the Dot product
```

```python
c = np.array([5,10,20])

d = np.array([18,32,50])

c_cross_d = np.cross(c,d)
```

```python
print(c_cross_d)
```

Projection of u vector on v vector

```python
u = np.array([2,5])

v = np.array([8,-6])

# magnitude of v vector

magnitude_v = np.sqrt(sum(v**2))

proj_u_on_v = (np.dot(u,v)/magnitude_v**2)*v

print("Projection of u vector on v vector ---->",proj_u_on_v)
```

```python
# For the 3D vectors
u = np.array([24,45,65])

v = np.array([76,98,34])

# magnitude of v vector

magnitude_v = np.sqrt(sum(v**2))

proj_u_on_v = (np.dot(u,v)/magnitude_v**2)*v

print("Projection of u vector on v vector ---->",proj_u_on_v)
```
