```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme()
```

```python
# Ploting a vector

plt.quiver(0,0,4,5)
plt.show()
```

```python
plt.quiver(0,0,4,5, scale_units='xy', angles='xy',scale=1,color='red')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,4,5, scale_units='xy', angles='xy',scale=1,color='b')
plt.quiver(0,0,-3,-6, scale_units='xy',angles='xy',scale=1,color='y')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
# Addition of Two Vectors
vector_1 = np.asarray([0,0,2,3])

vector_2 = np.asarray([0,0,3,-2])

sum = vector_1 + vector_2

print(sum)
```

```python
# Let's plot the Vector of above Calculation

plt.quiver(0,0,2,3, scale_units='xy', angles='xy',scale=1,color='red')
plt.quiver(0,0,3,-2, scale_units='xy',angles='xy',scale=1,color='green')
plt.quiver(0,0,5,1, scale_units='xy',angles='xy',scale=1,color='yellow')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
# Subraction of Two Vectors
#The subtraction of two vectors is defined as the addition of the first vector to the negative of the second
vector_1 = np.asarray([0,0,2,3])

vector_2 = np.asarray([0,0,3,-2])

sub = vector_1 - vector_2

print(sub)
```

```python
#Let's plot the graph
plt.quiver(0,0,2,3, scale_units='xy', angles='xy',scale=1,color='red')
plt.quiver(0,0,3,-2, scale_units='xy',angles='xy',scale=1,color='green')
plt.quiver(0,0,-1,5, scale_units='xy',angles='xy',scale=1,color='yellow')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
# Muliptlying a vector by scalar

vector_1 = np.asarray([0,0,2,3])

vector_2 = 2*vector_1
print(vector_2)  # Output: [0 0 4 6]
```

```python
# Let's plot the graph

plt.quiver(0,0,2,3, scale_units='xy', angles='xy',scale=1,color='red')
plt.quiver(0,0,4,6, scale_units='xy', angles='xy',scale=1,color='green')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
vector_1 = np.asarray([0,0,2,3])

vector_2 = -0.5*vector_1
print(vector_2)
```

```python
plt.quiver(0,0,2,3, scale_units='xy', angles='xy',scale=1,color='red')
plt.quiver(0,0,-1,-1.5, scale_units='xy', angles='xy',scale=1,color='green')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```
