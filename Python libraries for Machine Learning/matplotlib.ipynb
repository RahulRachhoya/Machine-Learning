```python
import matplotlib.pyplot as plt
```

```python
# import numpy to get the data 
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

```python
# Plotting the data
plt.plot(x,y)
plt.show
```

```python
# Cosine Wave
# The cosine wave is a periodic function that oscillates between 1 and -1. It is defined

plt.plot(x,z)
plt.show()
```

```python
# adding the title, x-axis & y-axis lables
plt.title('Plotting a sine wave')
plt.xlabel('angle')
plt.ylabel('sine value')
plt.plot(x,y)
plt.show()
```

```python
# Parabola 

plt.title('Plotting a parabola')
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
# I want to plot the parabola with '+' symbol and color is red

plt.title('Plotting a parabola')
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y, 'r+')
plt.show()
```

```python
# Let's do something more intresting let's plot the parabola in green symbol and in green color

plt.title('Plotting a parabola')
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y, 'g.')
plt.show()
```

```python
x = np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g-')

plt.plot(x,np.cos(x),'r--')

plt.show()
```

```python
# bar plots
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

langs = ['Python', 'C++', 'Java', 'JavaScript', 'Ruby']

students = [20, 15, 30, 25, 10]

ax.bar(langs, students)
plt.xlabel('languages')

plt.ylabel('students')

plt.show()
```

```python
# Pie Chart

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

langs = ['Python', 'C++', 'Java', 'JavaScript', 'Ruby']

students = [20, 15, 30, 25, 10]

ax.pie(students, labels = langs, autopct='%1.2f%%')

plt.show()
```

```python
# Scatter Plot

x = np.linspace(0,10,20)
y = np.sin(x)
z = np.cos(x)

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.scatter(x,y, color='r')

ax.scatter(x,z, color='g')

plt.show()
```

```python
# Built the 3D scatter plot
fig = plt.figure()
ax = plt.axes(projection='3d')
z = 20*np.random.random(100)
x = np.sin(z)

y = np.cos(z)

ax.scatter3D(x, y, z, c=z, cmap='viridis')

plt.show()
```
