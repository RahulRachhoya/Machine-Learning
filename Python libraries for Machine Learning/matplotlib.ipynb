Matplotlib : 
 usefull for making plots
 

```python
import matplotlib.pyplot as plt
```

```python
# import numpy to get the data 
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

```python
# Plotting the data
plt.plot(x,y)
plt.show
```

```python
# Cosine Wave
# The cosine wave is a periodic function that oscillates between 1 and -1. It is defined

plt.plot(x,z)
plt.show()
```

```python
# adding the title, x-axis & y-axis lables
plt.title('Plotting a sine wave')
plt.xlabel('angle')
plt.ylabel('sine value')
plt.plot(x,y)
plt.show()
```

```python
# Parabola 

plt.title('Plotting a parabola')
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.show()
```
