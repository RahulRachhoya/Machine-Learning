Seaborn library
 A Data Visualization Library

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Seaborn has some inbulit dataset
# Total bill vs tip dataset

tips = sns.load_dataset('tips')
print(tips.head())
```

```python
# Setting the theme for the plot

sns.set_theme()
```

```python
# Visualize the data
sns.relplot(data=tips,x='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```

```python
# loading the iris dataset

iris = sns.load_dataset('iris')
print(iris.head())
```

```python
# Scatter plot 

sns.scatterplot(data=iris,x='sepal_length',y='petal_length',hue='species')
```

```python
sns.scatterplot(data=iris,x='sepal_length',y='petal_width',hue='species')
```

```python
# Loading the titatinc data set

titanic = sns.load_dataset('titanic')
titanic
```

```python
# Count Plot

sns.countplot(data=titanic,x='survived',hue='sex')
```
