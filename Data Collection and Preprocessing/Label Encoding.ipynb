```python
# Label Encoding----> converting the label into numerical form
# importing the dependcies 
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
# Label Encoding of the Brest Cancer DataSet
# ===============================================

# load the dataset

cancer_data = pd.read_csv('Data/data.csv')
```

```python
# First Five Rows of the Data set

cancer_data.head()
```

```python
# Finding the Count of Different Labels

cancer_data['diagnosis'].value_counts()
```

```python
# Load the Label Encoder Function

label_encoder = LabelEncoder()
```

```python
label = label_encoder.fit_transform(cancer_data.diagnosis)
```

```python
# Appending the labels to data frame

cancer_data['target'] = label
```

```python
cancer_data.head()
```

```python
# 0----> Bengin
# 1----> Malignant

cancer_data['target'].value_counts()
```

Label Encoding of Iris Data

```python
# Loading the data set

iris_data = pd.read_csv('Data/iris_data.csv')

# First Five Rows of the Data set

iris_data.head()
```

```python
# Count the Values

iris_data['Species'].value_counts()
```

```python
# Loading the Label Encoder

label_encoder1 = LabelEncoder()

iris_label = label_encoder1.fit_transform(iris_data.Species)
```

```python
iris_data['target_Species'] = iris_label
```

```python
iris_data['target_Species'].value_counts()
```

```python
# 0-----> Iris-setosa
# 1-----> Iris-versicolor
# 2-----> Iris-virginica 

iris_data.head()
```

Conclusion 

In both the Brest Cancer and Iris Datasets, the Label Encoding technique was applied to convert the categorical labels into numerical form. This allowed for easier analysis and model building. The numerical representation of the labels in both datasets is as follows:

- Brest Cancer Dataset: 0 for Bengin and 1 for Malignant

- Iris Data Set: 0 for Iris-setosa, 1 for Iris-versicolor, and 2 for Iris-virginica.

This technique can be applied to any dataset where categorical labels need to be converted into numerical form.

Note: This is a simple example and in real-world scenarios, you might need to consider more complex preprocessing techniques, such as handling missing values, encoding categorical variables, and normalizing numerical features. These techniques can significantly improve the performance of your machine learning models.

Note: The LabelEncoder class from the sklearn.preprocessing module is used for this task. Other encoders like OneHotEncoder or LabelBinarizer can also be used depending on the specific requirements.

The actual implementation of these techniques can vary based on the specific dataset and the machine learning algorithm you are using. It's always a good idea to explore and experiment with different preprocessing techniques to find the best one for your specific use case.
