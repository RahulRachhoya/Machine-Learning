```python
# Importing the Libraries
import numpy as np
import pandas as pd
```

```python
# Loading the data set into pandas data frame

credit_card_df = pd.read_csv('Data/credit_data.csv')

# Displaying the first few rows of the data

credit_card_df.head()
```

```python
# distribution of the two classes 

credit_card_df['Class'].value_counts()
```

```python
# This is an Highly Imbalanced Data set
# 0--> Legit Transaction
# 1--> Fraudulent Transaction

# Seprating the legit and Fradulent Transaction

legit_df = credit_card_df[credit_card_df['Class']==0]
fraud_df = credit_card_df[credit_card_df['Class']==1]  
```

```python
print(legit_df.shape)

print(fraud_df.shape)
```

```python
# UnderSampling  -> Building a sample data set Containing the similar distrubution of Legit and Fradulent Transaction

legit_sample = legit_df.sample(n=492)
```

```python
print(legit_sample.shape)
```

```python
# Concat the Two data Frames

sampled_data = pd.concat([legit_sample, fraud_df],axis=0)
```
